# 1 - Atividade de Seleção dos Dados Brutos

```python
import pandas as pd
import numpy as np
```

```python
base_bruta = pd.read_csv('DADOS_IBOV_JAN_2020_2021.csv', sep=';', encoding='utf-8')
```

```python
base_bruta.shape
```

```python
base_bruta.dtypes
```

```python
d = base_bruta.set_index('Data')
```

```python
for x, y in enumerate (d.columns):
    print(x, y)
```

```python
#só quero os preços de fechamento - agora começa a seleção

d_close = d.iloc[:, 2:75]
```

```python
d_close.shape
```

```python
d_close = pd.concat([d['LastPrice'],d_close], axis=1) #axis =1 o valor vai lado a lado, não em baixo um do outro
```

```python
d_close
```

```python
d_vol = d.iloc[:, 75:147]
```

```python
d_vol = pd.concat([d['Volume'], d_vol], axis=1)
```

```python
d_vol
```

```python
d_prob = d.iloc[:, 147:161]
```

```python
d_prob = pd.concat([d['LastPrice'], d_prob],axis=1)
```

```python
d_prob
```

```python
d_hurst = d.iloc[:, 665:737]
```

```python
d_hurst = pd.concat([d['LastPrice'], d_hurst], axis=1)
```

```python
d_hurst
```

```python
d_power = d.iloc[:, 737:]
```

```python
d_power = pd.concat([d['LastPrice'], d_power], axis=1)
```

```python
d_power
```

Salvar dados selecionados

```python
d_close.to_csv('D_CLOSE.csv')
```

```python
d_vol.to_csv('D_VOL.csv')
d_hurst.to_csv('D_HURST.csv')
d_power.to_csv('D_POWER.csv')
d_prob.to_csv('D_PROB.csv')
```
